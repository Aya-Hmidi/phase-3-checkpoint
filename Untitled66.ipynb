```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("ola_ride_requests.csv")
```

```python
df.columns = ['user_ID', 'request_latitude', 'request_longitude', 'request_Time',
              'pickup_lat', 'pickup_long', 'drop_lat', 'drop_long']
```

```python
from geopy.distance import geodesic

df['distance'] = df.apply(lambda row: geodesic(
    (row['request_latitude'], row['request_longitude']),
    (row['pickup_lat'], row['pickup_long'])).km, axis=1)
```

```python
X = df[['request_latitude', 'request_longitude', 'distance']]
y_lat = df['pickup_lat']
y_long = df['pickup_long']
```

```python
X_train, X_test, y_lat_train, y_lat_test = train_test_split(X, y_lat, test_size=0.2, random_state=42)
_, _, y_long_train, y_long_test = train_test_split(X, y_long, test_size=0.2, random_state=42)
```

```python
param_grid = {
    'n_estimators': [100, 150],
    'max_depth': [10, 15],
    'min_samples_split': [2, 3],
    'min_samples_leaf': [1, 2]}
```

```python
grid_long = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=3, scoring='r2')
grid_long.fit(X_train, y_long_train)
best_long_model = grid_long.best_estimator_
```

```python
grid_long = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=3, scoring='r2')
grid_long.fit(X_train, y_long_train)
best_long_model = grid_long.best_estimator_
```

```python
y_lat_pred = best_lat_model.predict(X_test)
y_long_pred = best_long_model.predict(X_test)
```

```python
r2_lat = r2_score(y_lat_test, y_lat_pred)
mse_lat = mean_squared_error(y_lat_test, y_lat_pred)
print("\n Latitude Prediction (Tuned Random Forest)")
print("R² Score:", r2_lat)
print("MSE:", mse_lat)
```

```python
r2_long = r2_score(y_long_test, y_long_pred)
mse_long = mean_squared_error(y_long_test, y_long_pred)
print("\n Longitude Prediction (Tuned Random Forest)")
print("R² Score:", r2_long)
print("MSE:", mse_long)
```

```python
print("\n Best Parameters (Latitude):", grid_lat.best_params_)
print(" Best Parameters (Longitude):", grid_long.best_params_)
```

```python

from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error

xgb_param_grid = {
    'n_estimators': [100, 150],
    'max_depth': [5, 10, 15],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}


xgb_lat = GridSearchCV(
    XGBRegressor(random_state=42, objective='reg:squarederror'),
    xgb_param_grid,
    cv=3,
    scoring='r2',
    verbose=1,
    n_jobs=-1
)
xgb_lat.fit(X_train, y_lat_train)
best_xgb_lat = xgb_lat.best_estimator_
y_lat_pred_xgb = best_xgb_lat.predict(X_test)

xgb_long = GridSearchCV(
    XGBRegressor(random_state=42, objective='reg:squarederror'),
    xgb_param_grid,
    cv=3,
    scoring='r2',
    verbose=1,
    n_jobs=-1
)
xgb_long.fit(X_train, y_long_train)
best_xgb_long = xgb_long.best_estimator_
y_long_pred_xgb = best_xgb_long.predict(X_test)

print("Latitude (XGBoost):")
print("R² Score:", r2_score(y_lat_test, y_lat_pred_xgb))
print("MSE:", mean_squared_error(y_lat_test, y_lat_pred_xgb))
print("Best Params:", xgb_lat.best_params_)

print("\n Longitude (XGBoost):")
print("R² Score:", r2_score(y_long_test, y_long_pred_xgb))
print("MSE:", mean_squared_error(y_long_test, y_long_pred_xgb))
print("Best Params:", xgb_long.best_params_)

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(y_lat_test, y_lat_pred_xgb, alpha=0.5, color='green')
plt.plot([min(y_lat_test), max(y_lat_test)], [min(y_lat_test), max(y_lat_test)], 'r--')
plt.xlabel('True Latitude')
plt.ylabel('Predicted Latitude')
plt.title('XGBoost - Latitude Prediction')

plt.subplot(1, 2, 2)
plt.scatter(y_long_test, y_long_pred_xgb, alpha=0.5, color='orange')
plt.plot([min(y_long_test), max(y_long_test)], [min(y_long_test), max(y_long_test)], 'r--')
plt.xlabel('True Longitude')
plt.ylabel('Predicted Longitude')
plt.title('XGBoost - Longitude Prediction')

plt.tight_layout()
plt.show()
```
